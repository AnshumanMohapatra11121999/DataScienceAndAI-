```python
#Types of method-
#1.class methods
'''
.Bound to class
.Receive class object as the implicit first argument
.'cls' parameter represents the class object
.cannot access instance variables
.defined using @classmethod

'''

#2.static method:

'''
.Bound to class
.Doesnt require any implicit first argument
.Implementing functionality having logical connection
.cannot access instance variables
.Defined using @static method

'''
```

```python
class Circle2:
  pi=3.14
  def __init__(self,radius):
    self.radius = radius

#Instance Method
  def calculate_area(self):
   return Circle2.pi * self.radius

#Class Method
  @classmethod
  def access_pi(cls):
   cls.pi=3.1436
   return cls.pi

#static Method
  @staticmethod
  def circle_static_method():
   print("This is a circles static method")
```

```python
#CREATE AN OBJECT
cir=Circle2(5)
```

```python
cir.calculate_area()#static method is called by instance of the class
```

```python
#Class method can modify the variable inside the class
Circle2.pi
```

```python
cir.pi
```

```python
Circle2.access_pi()
```

```python
#Now check the value of pi
cir.pi
```

```python
#To call a static method we need to call it by class name
Circle2.circle_static_method()
```

```python
#Inheritance and Overriding
```

```python
#Inheritance
'''
Inheritance helps in code reusability.
Just like a child inherits their parents' qualities,
the class that inherits properties is known as a child class,
and the class from which it inherits the properties is called the parent class.

'''
```

```python
class Shape:
  def set_colour(self,colour):
   self.colour=colour
  def calculate_area(self):
    pass
  def colour_the_shape(self):
    colour_price={"red":10,"blue":15,"green":5}
    return self.calculate_area()*colour_price[self.colour]
```

```python
class Circle3(Shape):#This type of class declaration means circle 3 class inherits Shape class
  pi=3.14
  def __init__(self,radius):
    self.radius=radius

  #Example of overriding
  def calculate_area(self):
    return Circle3.pi*self.radius
```

```python
c=Circle3(5)
c.set_colour("red")#here we are using the method of the parent class using the object of child class.
#This is the example of inheritance.
print("Circle with radius = ",c.radius,"when coloured",c.colour,"costs rupees",c.colour_the_shape())
```

```python
class Rectangle(Shape) :
  def __init__(self,length,bredth):
    self.length=length
    self.bredth=bredth

  #Overriding user defined method
  def calculate_area(self):
    return self.length*self.bredth

  #Overriding python default method
  def __str__(self):
    return "area of rectangle = "+str(self.Calculate_area())
```

```python
r=Rectangle(5,10)
r.set_colour("blue")
print("Rectangle with length =",r.length," and bredth = ",r.bredth,"When coloured ",r.colour,"costs rupees",r.colour_the_shape())
```
