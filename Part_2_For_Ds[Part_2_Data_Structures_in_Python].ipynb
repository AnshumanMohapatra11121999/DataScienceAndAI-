```python
#Difference between sort and sorted
```

```python
'''
1.sort():
Method of a list object.

Modifies the list in place (changes the original list).

Returns None.

Only works on lists.

2.sorted()
Built-in function.

Returns a new sorted list, leaving the original unchanged.

Can work on any iterable (list, tuple, string, etc.).

,,,
```

```python
#example
listsort=[1,2,3,4,5]
listsort.sort()
print(listsort)
#Sort the list in desecending order
listsort.sort(reverse=True)
print(listsort)

```

```python
a=["org","app","Mango"]
b=a.sort()#It sorts the list according to the ascii value and returns nothing it sorts the original value(the sort())
print(a)
print(b)
```

```python
#But when we are using sorted it returns a new list
c=sorted(a)
print(a)
print(c)
```

```python
#Shallow copying
x=[1,2,3]
y=x
x[0]=0#we only changed in list x not in y
print(x)
print(y)
#still the change is visible in y too.because when = is used to assign values of x to y it just refers the address of the original list
```

```python
#To over come the above problem we use shallow copying.Example given below
```

```python
z=x[:]#we will use end to end slicing then lets see if we change any thing in x does it refects in z too
```

```python
x[2]=2
print(x)
print(z)
```

```python
#As you can see it doesnt make any changes
```

```python
#Implement set operations on the sets
#Union,intersection,difference,symmetric difference
```

```python
A={0,1,2,3,4,5}
B={4,5,6,7,8,9}
```

```python
print(A.union(B))
#or we can write like this
print(A|B)
#Both does the union operation
```

```python
print(A.intersection(B))
#or we can write like this
print(A&B)
#Both does the intersection operation
```

```python
print(A.difference(B))
#or we can write like this
print(A-B)
#Both does the difference operation
```

```python
print(A.symmetric_difference(B))
#or we can write like this
print(A^B)
#Both does the Symmetric difference operation

#The symmetric difference of two sets is the set of elements that are in either of the sets,
#but not in both.
```

```python
(A.union(B)).difference(A.intersection(B))
```
