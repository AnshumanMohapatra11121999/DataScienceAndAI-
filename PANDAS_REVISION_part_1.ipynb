```python
#Another method to handle missing value is replacing the null value with mean,median and mode
#If the data is normal then we can go with mean()
#If the data consists special value such as outliers then we can go with median()
#If the data is categorical(such as male,female) we can go with the majority that means mode()
```

```python
#So the second most nana values were in solar colum lets replace it eith mean
print(df["Solar.R"].isnull().sum())#7 null values were there
#lets calculate the mean and replace it with the null values present in the solar.r column
mean1=df["Solar.R"].mean()
print(mean1)
#lets replace it with the null values in the solar r column
df["Solar.R"]=df["Solar.R"].fillna(mean1)
print(df["Solar.R"].isnull().sum())#Now if we check there is no null value in the solar.r column
```

```python
#We need to remove the null values from weather column
print(df["Weather"].isnull().sum())
#As the value present here in this column is categorical("NON-numerical")so we can impute/replace it with the mode of the all the values present in the row
print(df["Weather"].mode())
#lets replace the mode with the missing value with the mode off the values in the weather column
mode1=df["Weather"].mode()[0]#we have used [0] because the returned type by Mode() method is series.so we need to index 0
df["Weather"]=df["Weather"].fillna(mode1)
#Now the missing value is replaced
print(df["Weather"].isnull().sum())
```

```python
#CONCAT AND APPEND
import pandas as pd

# Create DataFrames 1 and 2 based on the image content
df1 = pd.DataFrame({
    'Name': ['Shyam', 'Ram', 'Radha'],
    'Age': [30, 25, 23],
    'Gender': ['Male', 'Male', 'Female']
})

df2 = pd.DataFrame({
    'Name': ['Mark'],
    'Age': [27],
    'Gender': ['Male']
})



```

```python
df1
```

```python
df2
```

```python
#Concatenate dataframes vertically(STACKING ROWS)
df_concat=pd.concat([df1,df2],ignore_index=True)#Reset index to avaoid duplicate
df_concat
```

```python
#Create a new row as a Series
new_row=pd.Series({"Name":"Juhi","Age":'32',"Gender":'Female'})
dfconcatagain=pd.concat([df_concat,new_row],ignore_index=True)

```

```python
dfconcatagain#here the null values comes if we try to add the series as a new row in the existing data frame and it enters everything in a same column
```

```python
#We can use append here to add the row in the existing format
dfconcat_append=df_concat.append(new_row,ignore_index=True)

'''
âŒ DataFrame.append() is REMOVED

In pandas 2.0+, DataFrame.append() does not exist

Using it raises:

AttributeError: 'DataFrame' object has no attribute 'append'


âœ… Use pd.concat() instead




'''
```

```python
# -------------------------------
# Issue Explanation: Series â†’ DataFrame separately
# -------------------------------

# When you do this:
new_row_asdf = pd.DataFrame(new_row)
print(type(new_row_asdf))  # <class 'pandas.core.frame.DataFrame'>

# And then concatenate:
dfconcat_2 = pd.concat([df_concat, new_row_asdf], ignore_index=True)
print(dfconcat_2)

# Output:
#    Name   Age  Gender       0
# 0  Shyam  30.0    Male     NaN
# 1    Ram  25.0    Male     NaN
# 2  Radha  23.0  Female     NaN
# 3   Mark  27.0    Male     NaN
# 4    NaN   NaN     NaN    Juhi
# 5    NaN   NaN     NaN      32
# 6    NaN   NaN     NaN  Female

# -------------------------------
# Why this happened:
# -------------------------------
# 1. new_row is a Series with index: ["Name", "Age", "Gender"]
# 2. pd.DataFrame(new_row) converts the Series into a **single-column DataFrame**
#    where the Series index becomes the **row index** and the Series values become the **column values**
# 3. The resulting DataFrame has **column named 0**, not matching df_concat columns
# 4. When concatenated, pandas aligns **columns by name**:
#    - Existing columns ("Name","Age","Gender") â†’ get NaN
#    - The Series values go into the **new column 0**
# 5. Result: extra column created and original columns are NaN in the new rows

# -------------------------------
# Key Note / Rule:
# -------------------------------
# âœ… To add a Series as a new row correctly:
# - Wrap it in a list inside pd.DataFrame inside concat:
#   pd.concat([df_concat, pd.DataFrame([new_row])], ignore_index=True)
# - This ensures **columns match** and no extra column is created
```

```python
# -------------------------------
# Add a Series as a new row without creating extra columns
# -------------------------------

# df_concat : existing DataFrame
# new_row   : Series representing the new row

# Convert Series to DataFrame inside concat to match columns properly
dfconcat_3 = pd.concat([df_concat, pd.DataFrame([new_row])], ignore_index=True)

# Check the result
dfconcat_3  # Worked properly

# -------------------------------
# NOTE:
# While adding a Series as a row to an existing DataFrame,
# always convert the Series to a DataFrame inside pd.concat()
# to avoid creating extra columns with NaN values
# -------------------------------
```

```python
df3.groupby(["sex","rank"]).agg({"rank":"count","rank":"nunique"})
# In agg(), dictionary keys must be unique; using "rank" twice causes overwrite
# {"rank":"count","rank":"nunique"} becomes {"rank":"nunique"}
# Therefore "rank":"count" is ignored and never executed
# groupby(["sex","rank"]) creates groups with only one rank per group
# nunique(rank) returns 1 because each group contains a single unique rank
# This is why all output values are 1
# To apply multiple aggregations, use named aggregations or a list of functions
# Correct example: agg(count=("rank","count"), unique=("rank","nunique"))
# If only row counts are needed, use groupby(...).size()
# Memory tip: applying nunique on a grouped column always returns 1
```

```python
#CHEAT SHEET FOR PANDAS DATAFRAME:
'''
1. Creating & Inspecting DataFrames

pd.DataFrame() â€” Create a DataFrame

.head(n) â€” View first n rows

.tail(n) â€” View last n rows

.info() â€” Summary of DataFrame structure and types

.describe() â€” Statistical summary of numeric columns

.shape â€” Get (rows, columns) tuple

.columns â€” List column names

.dtypes â€” Data types of each column

2. Selecting & Filtering Data

df["col"] or df.col â€” Select a column (Series)

df[["col1", "col2"]] â€” Select multiple columns (DataFrame)

df.loc[row_labels, col_labels] â€” Label-based selection

df.iloc[row_positions, col_positions] â€” Position-based selection

df[df["col"] > value] â€” Filter rows based on condition

df.query("col > value") â€” Query using string expressions

3. Adding & Modifying Columns

df["new_col"] = values â€” Add or replace column

.assign(new_col=values) â€” Add column (returns new DataFrame)

.astype(type) â€” Change data type of columns

4. Handling Missing Data

.isnull() / .isna() â€” Detect missing values

.notnull() / .notna() â€” Detect non-missing values

.dropna() â€” Remove rows/columns with missing data

.fillna(value) â€” Replace missing values

5. Sorting

.sort_values(by="col", ascending=True) â€” Sort by column(s)

.sort_index() â€” Sort by row index

6. Grouping & Aggregating

.groupby(["col1", "col2"]) â€” Group data by column(s)

.agg({"col": "sum"}) â€” Aggregate with functions like sum, mean, count, etc.

.size() â€” Count rows in each group

.mean(), .sum(), .count(), .min(), .max(), .median() â€” Aggregate functions

7. Merging & Joining

pd.merge(df1, df2, on="key", how="inner") â€” Merge DataFrames on key(s)

.join() â€” Join by index or column

.concat([df1, df2], axis=0 or 1) â€” Concatenate DataFrames vertically or horizontally

8. Reshaping

.pivot(index, columns, values) â€” Pivot data from long to wide format

.melt() â€” Unpivot from wide to long format

.stack() / .unstack() â€” Reshape hierarchical indices

.transpose() or .T â€” Transpose rows and columns

9. Applying Functions

.apply(func) â€” Apply function along rows or columns

.applymap(func) â€” Apply function elementwise (for DataFrame)

.map(func) â€” Map values (for Series)

.replace() â€” Replace values with others

10. Datetime Operations

Convert to datetime: pd.to_datetime()

.dt accessor for datetime properties (year, month, day, etc.)

.resample() â€” Resample time series data

11. Exporting & Importing Data

pd.read_csv() â€” Read CSV files

df.to_csv() â€” Save DataFrame to CSV

pd.read_excel() / df.to_excel() â€” Excel files

pd.read_json() / df.to_json() â€” JSON format

12. Indexing & Resetting

.set_index("col") â€” Set column as index

.reset_index() â€” Reset index to default integers

.rename(columns={"old": "new"}) â€” Rename columns

13. Miscellaneous

.copy() â€” Create a copy of the DataFrame

.unique() â€” Find unique values in a Series

.value_counts() â€” Count unique values in a Series

.duplicated() / .drop_duplicates() â€” Detect or remove duplicates

Bonus: Common Aggregate Functions

count â€” Number of non-null values

nunique â€” Number of unique values

mean, median â€” Average/median

sum, min, max â€” Sum/minimum/maximum values

std, var â€” Standard deviation and variance '''
```
