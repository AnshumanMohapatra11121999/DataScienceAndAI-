```python

#Angram check
x=input()
y=input()
def anagram(x,y):
  if(len(x)!=len(y)):
    return False
  for i in range(len(x)):
    if(x.count(x[i])!=y.count(x[i])):
      return False
  return True

print(anagram(x,y))
```
