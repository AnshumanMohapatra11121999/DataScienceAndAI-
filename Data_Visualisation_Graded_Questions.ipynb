### II) Superstore Data

In the `superstore.csv` file, you have the details of orders purchased in an American online retail store. Load the dataset, observe and analyse the different columns and answer the following questions.

```python
#Load the dataset
df2 =pd.read_csv("/content/superstore.csv")
```

```python
df2
```

```python
#Answer for question 2
df3 = df2[(df2.Profit < 0) & (df2.Sales < 15000)]
sns.jointplot(x='Sales',y='Profit', data=df3)
plt.show()
```

```python
#Data Visualisation
#Ship Mode column has 4 categories – First Class, Second Class, Same Day and Standard Class, while the Segment has 3 categories – Consumer, Home Office and Corporate.
#Which of the following can be used to visualise how the average Profit varies across every Segment - Ship Mode combination (like Consumer - First Class, Corporate - Standard Class and so on)?
```

```python
pt = df2.pivot_table(values="Profit",
                     index="Segment",
                     columns="Ship Mode",
                     aggfunc="mean")

sns.heatmap(pt, annot=True, fmt=".0f")
plt.show()
```

```python

#Q3
'''
Bar Plot
Which of the following codes would help you create a bar plot in seaborn that compares the average Sales across the different Segment types?'''
```

```python
sns.barplot(data = df2, x = 'Segment', y = 'Sales', estimator = np.mean)
```

```python
sns.barplot(data = df2, x = 'Segment', y = 'Sales')
```
