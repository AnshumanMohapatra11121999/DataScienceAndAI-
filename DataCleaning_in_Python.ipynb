Getting Data From APIs

APIs, or application programming interfaces, are created by companies and organisations to provide restricted access to data. It is very common to get data from APIs for data analysis; for example, you can get financial data (stock prices, etc.), social media data (Facebook, Twitter, etc.), weather data, data about healthcare, music, food and drinks, and from almost every domain.



Apart from being rich sources of data, there are other reasons to use APIs:

1.We use APIs when the data requires to be updated in real-time. By contrast, if you use downloaded CSV files, you'll have to download data manually every time it changes and update your analysis multiple times in your program. Through APIs, you can automate the process of getting real-time data.

2.APIs also provide easy access to structured and verified data. Though you can scrape websites, APIs can directly offer data in a structured format with better quality.
3.APIs provide access to restricted data. Often, you cannot scrape all websites easily, which is often illegal (Example: Facebook, financial data, etc.). APIs are the only way to get this data.

```python
# requests = library to talk to URLs (APIs or websites)
import requests

# This is an API URL (returns data, not HTML)
url = "https://httpbin.org/get"

# Send a GET request to the API
# Think: "Hey server, give me the data at this URL"
response = requests.get(url, timeout=20)

# ALWAYS check status code first
# 200 = success
print("Status code:", response.status_code)

# Content-Type tells you what kind of data came back
# application/json = API
print("Content-Type:", response.headers.get("Content-Type"))

# Raw text response (just to visually confirm it's JSON)
print("Raw response (first 300 chars):")
print(response.text[:300])

#URL  ‚Üí  server  ‚Üí  response (JSON text)
```

```python
# Convert JSON text into Python objects (dict + list)
data = response.json()

# APIs almost always return:
# - dict (key-value)
# - OR list of dicts
print("Type of data:", type(data))

# Top-level keys tell you what data exists
print("Top-level keys:", list(data.keys()))

# Print full JSON (safe here because response is small)
print("\nFull JSON:")
print(data)
#JSON  ‚Üí  Python dict/list
```

```python
#Read JSON step-by-step (NO guessing)
# Rule: dict ‚Üí use keys, list ‚Üí use index

# Step 1: Access "args"
# args = query parameters sent to the API
args = data["args"]

print("args type:", type(args))
print("args content:", args)

```

```python
# "headers" is another dictionary inside the JSON
headers = data["headers"]

print("headers type:", type(headers))
print("header keys:", list(headers.keys()))
```

```python
#Safely extract values from JSON
# Use .get() to avoid KeyError if key doesn't exist
user_agent = headers.get("User-Agent")
host = headers.get("Host")

print("User-Agent:", user_agent)
print("Host:", host)
```

```python
# APIs use ONLY dicts and lists
print("Is headers a list?", isinstance(headers, list))
print("Is headers a dict?", isinstance(headers, dict))
```

```python
#Send query parameters to an API
# Query parameters are sent as a dictionary
params = {
    "name": "alice",
    "skill": "python"
}

# requests automatically builds the URL for you
response = requests.get(url, params=params, timeout=20)

# Convert response to JSON again
data = response.json()

# The server echoes back what it received
print("Query parameters received by server:")
print(data["args"])

```

```python
#See the final URL (proof)
# This shows the actual URL that was called
print("Final URL used:")
print(data["url"])
```

```python
#Recognize API vs Website
# Website (HTML)
website_response = requests.get("https://example.com", timeout=20)

# API (JSON)
api_response = requests.get("https://httpbin.org/get", timeout=20)

print("Website response starts with:")
print(website_response.text[:40])

print("\nAPI response starts with:")
print(api_response.text[:40])
```

Reading Data From PDF Files

1.Reading PDF files is not as straightforward as reading text or delimited files using Python since PDFs often contain images, tables, etc. PDFs are mainly designed to be human-readable, and thus you need special libraries to read them in Python (or any other programming language).

2.There are some great libraries in Python. We will use PyPDF2 to read PDFs in Python since it is easy to use and works with most PDFs.

3.Python will only be able to read text from PDFs, not images, tables etc (though that is possible using other specialised libraries).

```python
%pip install PyPDF2
```

```python
from PyPDF2 import PdfReader

# Path to your PDF file
pdf_path = "/content/animal_farm.pdf"

# Create a PDF reader object
reader = PdfReader(pdf_path)

# Number of pages in the PDF
print("Total pages:", len(reader.pages))
```

```python
#Read text from one page
# Get the first page (index starts at 0)
page = reader.pages[3]

# Extract text from the page
text = page.extract_text()

print(text)
```

```python

#Read all pages
all_text = ""

for i, page in enumerate(reader.pages):
    page_text = page.extract_text()

    print(f"--- Page {i+1} ---")
    print(page_text[:300])  # preview

    if page_text:
        all_text += page_text + "\n"
```

```python
#Save extracted text to a file
with open("pdfoutput.txt", "w", encoding="utf-8") as f:
    f.write(all_text)
```

Treating Missing Values
Missing values can be handled in one of two ways:

1.Delete: Remove the values that aren‚Äôt present.

2.Impute:Using a simple statistic to infer: Replace the missing values with another value, such as the mean, median, or mode.
Techniques for prediction: To predict and impute missing values, use statistical models such as kNN, SVM, and others.
Imputation, in general, makes assumptions about missing values and replaces them with arbitrary estimates such as mean, median, and so on. It should only be used if you are pretty certain about your assumptions.



Otherwise, deletion is usually the safest and more recommended option. You could lose some information, but you won‚Äôt make any irrational assumptions.

```python
#land size column has 9% null value lets check this
df["Landsize"].describe()
#optional
# NOTE: `count` represents the number of NON-missing values in the column
# NOTE: If count < total number of rows, missing values are present

# NOTE: Mean formula -> mean = (sum of values) / n
# NOTE: Max formula -> max = maximum observed value
# NOTE: If max >> mean, the distribution has a long right tail (right skew)

# NOTE: Median = 50th percentile (robust to outliers)
# NOTE: Rule: If mean > median, data is positively (right) skewed
# NOTE: In this column: mean (558) > median (440) confirms right skewness

# NOTE: Standard deviation formula -> std = sqrt( sum((x - mean)^2) / n )
# NOTE: Large std relative to mean indicates extreme variability
# NOTE: Here std (3987) >> mean (558), suggesting strong outliers

# NOTE: Min = 0 means zero exists in the data
# NOTE: Zero should NOT be treated as missing unless domain knowledge says so

# NOTE: IQR formula -> IQR = Q3 - Q1
# NOTE: Upper outlier bound = Q3 + 1.5 * IQR
# NOTE: Values far above this bound are considered extreme outliers
```

```python
#lets remove the rows which contains null values in the "Landsize "column
df=df[~np.isnan(df["Landsize"])]
```

```python
round(100*(df.isnull().sum()/len(df.index)),2)#lets check the percentage of missing value
```

```python
#We learnt how to handle missing values by deleting them. Now let's learn how to impute them with the mean value of the overall column
```

```python
df.loc[:,["Lattitude","Longtitude"]].describe()
```

```python
#As the std deviation is very less we can use mean imputation here
#imputing the latititidea and longtitude by mean values
df.loc[np.isnan(df["Lattitude"]),["Lattitude"]]=df["Lattitude"].mean()
df.loc[np.isnan(df["Longtitude"]),["Longtitude"]]=df["longtitude"].mean()
# ===================== NOTES: Boolean Indexing with loc & np.isnan =====================
#
# 1. df["Lattitude"] returns a pandas Series (NOT a DataFrame).
#    - A Series is 1-dimensional (only one axis).
#    - Hence, there is NO need to specify axis=0 or axis=1.
#
# 2. np.isnan(df["Lattitude"]) works element-wise on the Series
#    and returns a boolean Series:
#       True  -> value is NaN
#       False -> value is NOT NaN
#
# 3. The boolean Series has the SAME index as the DataFrame.
#    Pandas automatically aligns it with df.index.
#
# 4. When used inside df.loc[ boolean_condition , column ],
#    pandas interprets the boolean Series as a ROW filter.
#
#    Syntax:
#    df.loc[ROWS_TO_SELECT, COLUMNS_TO_UPDATE] = VALUE
#
# 5. Even if the same column name appears twice in loc[],
#    they serve DIFFERENT purposes:
#       - First occurrence  -> to identify rows (condition)
#       - Second occurrence -> to specify which column to assign values to
#
# 6. Axis is required ONLY for 2D objects (DataFrames),
#    but NOT for Series (1D objects).
#
# Example:
# df.loc[np.isnan(df["Lattitude"]), ["Lattitude"]] = df["Lattitude"].mean()
#
# Meaning:
# "For rows where Lattitude is NaN, replace Lattitude with its mean value"
#
# =======================================================================================
```
