```python
#lets learn to display the three graphs in a separet view inside a single graph using subplots
fig,ax=plt.subplots(ncols=3)
africa,=ax[0].plot(years,sales_Africa)
africa.set_label("Africa")
aspac,=ax[1].plot(years,sales_Asia_Pacific)
aspac.set_label("Asia-Pacific")
europe,=ax[2].plot(years,sales_Europe)
europe.set_label("Europe")
europe.set_dashes([2,2,5,2])
plt.legend()
plt.show()
# This code displays the three graphs in separate views (subplots)
# inside a single figure, instead of plotting all lines on one graph.
#
# fig, ax = plt.subplots(ncols=3)
# creates ONE figure with THREE side-by-side axes (columns).
#
# ax is an array of axes:
#   ax[0] -> Africa plot
#   ax[1] -> Asia-Pacific plot
#   ax[2] -> Europe plot
#
# Each region is plotted on its own axis:
#   ax[0].plot(...) plots Africa only
#   ax[1].plot(...) plots Asia-Pacific only
#   ax[2].plot(...) plots Europe only
#
# This is different from the previous code where all regions
# were plotted on the SAME axis, overlapping in one graph.
#
# Purpose / Advantage:
# - Makes each region easier to read and compare individually
# - Avoids overlapping lines and visual clutter
# - Useful when trends have different scales or patterns
#
# europe.set_dashes([2,2,5,2])
# applies a custom dashed line style to the Europe subplot.
#
# Note:
# plt.legend() will not work correctly for separate subplots.
# Each subplot needs its own legend:
#   ax[i].legend()
```

```python
# Import the pyplot module from matplotlib
# pyplot provides functions to create plots and charts
import matplotlib.pyplot as plt


# Sample data used for plotting
# x represents values on the x-axis
x = [1, 2, 3, 4, 5]

# y1 to y5 represent different datasets for different subplots
y1 = [1, 2, 3, 4, 5]
y2 = [5, 4, 3, 2, 1]
y3 = [2, 3, 2, 3, 2]
y4 = [1, 3, 1, 3, 1]
y5 = [2, 4, 6, 8, 10]


# Create a subplot in a 3x3 grid at position 1
# Position counting starts from top-left and moves left to right
plt.subplot(3, 3, 1)

# Plot y1 vs x with a solid line style
# label is used for the legend
plt.plot(x, y1, linestyle='-', label='Solid Line')

# Display the legend for this subplot
plt.legend()

# Set the title for subplot 1
plt.title("Subplot 1")


# Create subplot at position 2 (row 1, column 2)
plt.subplot(3, 3, 2)

# Plot y2 vs x with a dashed line style
plt.plot(x, y2, linestyle='--', label='Dashed Line')

# Show legend for subplot 2
plt.legend()

# Set the title for subplot 2
plt.title("Subplot 2")


# Create subplot at position 3 (row 1, column 3)
plt.subplot(3, 3, 3)

# Plot y3 vs x with a dotted line style
plt.plot(x, y3, linestyle=':', label='Dotted Line')

# Show legend for subplot 3
plt.legend()

# Set the title for subplot 3
plt.title("Subplot 3")


# Create subplot at position 4 (row 2, column 1)
plt.subplot(3, 3, 4)

# Plot y4 vs x with a dash-dot line style
plt.plot(x, y4, linestyle='-.', label='Dash-Dot Line')

# Show legend for subplot 4
plt.legend()

# Set the title for subplot 4
plt.title("Subplot 4")


# Create subplot at position 5 (row 2, column 2)
plt.subplot(3, 3, 5)

# Plot y5 vs x with a dashed line and increased thickness
# linewidth controls how thick the line appears
plt.plot(x, y5, linestyle='--', linewidth=2, label='Thick Dashed Line')

# Show legend for subplot 5
plt.legend()

# Set the title for subplot 5
plt.title("Subplot 5")


# Display all the subplots together in one figure
plt.show()
```
