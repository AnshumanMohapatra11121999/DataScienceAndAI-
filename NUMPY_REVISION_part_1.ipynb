```python
import numpy as np

# -------------------------------
# Shallow Copy vs Deep Copy in NumPy
# -------------------------------

m = np.array([1, 2, 3, 4, 5])
# Original NumPy array created

n = m
# n is NOT a new array
# n is just another reference to the same memory as m
# This behaves like a SHALLOW COPY (reference assignment)

o = np.copy(m)
# o is a DEEP COPY
# A new array with a new memory location is created
# Changes in m will NOT affect o

print("Before changes:")
print("m:", m)
print("n:", n)
print("o:", o)

# -------------------------------
# Modifying original array
# -------------------------------

m[0] = 9
# Changing first element of m

print("\nAfter changing m[0] = 9:")
print("m:", m)
# m reflects the change

print("n:", n)
# n ALSO reflects the change because it shares memory with m

print("o:", o)
# o does NOT change because it has separate memory
```

```python
# =========================
# NUMPY COMPLETE CHEAT SHEET
# (1D & 2D ARRAYS)
# SINGLE CELL WITH PRINTS
# =========================

import numpy as np

print("----- Creating 1D Array -----")
a = np.array([1, 2, 3, 4, 5])
print(a)
# 1D NumPy array created


print("\n----- Creating 2D Array -----")
b = np.array([[1, 2, 3],
              [4, 5, 6]])
print(b)
# 2D NumPy array created


print("\n----- Array Attributes -----")
print("a ndim:", a.ndim)
print("b ndim:", b.ndim)
print("a shape:", a.shape)
print("b shape:", b.shape)
print("a size:", a.size)
print("b size:", b.size)
print("a dtype:", a.dtype)
# Basic properties of arrays


print("\n----- Special Arrays -----")
z = np.zeros((2, 3))
print("Zeros:\n", z)

o = np.ones((2, 2))
print("Ones:\n", o)

i = np.eye(3)
print("Identity Matrix:\n", i)
# Zeros, ones, and identity matrix


print("\n----- Range Based Arrays -----")
r = np.arange(1, 10, 2)
print("Arange:", r)

l = np.linspace(0, 1, 5)
print("Linspace:", l)
# Creating arrays using range and spacing


print("\n----- Reshape vs Resize -----")
c = np.array([1, 2, 3, 4, 5, 6])
print("Original array c:", c)

reshaped_c = c.reshape(2, 3)
print("After reshape (2x3):\n", reshaped_c)
print("Original c after reshape:", c)
# reshape() DOES NOT change original array


d = np.array([1, 2, 3, 4, 5, 6])
print("\nOriginal array d:", d)

d.resize(2, 3)
print("After resize (2x3):\n", d)
# resize() CHANGES original array


e = np.array([1, 2, 3])
print("\nOriginal array e:", e)

e.resize(2, 3)
print("After resize (size increased):\n", e)
# resize() repeats elements if size increases


print("\n----- Flatten -----")
b_flat = b.flatten()
print("Flattened b:", b_flat)
# Converts 2D array to 1D


print("\n----- Indexing & Slicing -----")
print("First element of a:", a[0])
print("Last element of a:", a[-1])
print("Slice a[1:4]:", a[1:4])
print("Element at row 0 col 1 in b:", b[0, 1])
print("Second column of b:", b[:, 1])
# Accessing and slicing arrays


print("\n----- Vectorized Operations -----")
print("a + 10:", a + 10)
print("a * 2:", a * 2)
print("a squared:", a ** 2)
# Element-wise operations without loops


print("\n----- Universal Functions -----")
print("Square root:", np.sqrt(a))
print("Sum:", np.sum(a))
print("Mean:", np.mean(a))
print("Max:", np.max(a))
print("Min:", np.min(a))
print("Std Dev:", np.std(a))
# Mathematical and statistical functions


print("\n----- Axis Operations -----")
print("Column-wise sum of b:", np.sum(b, axis=0))
print("Row-wise sum of b:", np.sum(b, axis=1))
# Axis-based operations in 2D arrays


print("\n----- Boolean Indexing -----")
print("a > 3:", a > 3)
print("Filtered values:", a[a > 3])
# Filtering using conditions


print("\n----- Random Arrays -----")
np.random.seed(42)
rand1 = np.random.rand(5)
print("Random 1D:", rand1)

rand2 = np.random.randint(1, 10, size=(2, 3))
print("Random 2D:\n", rand2)
# Random number generation


print("\n----- Sorting & Searching -----")
print("Sorted a:", np.sort(a))
print("Index of max:", np.argmax(a))
print("Index of min:", np.argmin(a))
# Sorting and searching operations


print("\n----- Matrix Operations -----")
x = np.array([[1, 2],
              [3, 4]])
y = np.array([[5, 6],
              [7, 8]])

print("Matrix x:\n", x)
print("Matrix y:\n", y)
print("Addition:\n", x + y)
print("Multiplication:\n", x @ y)
print("Transpose:\n", x.T)
print("Determinant:", np.linalg.det(x))
print("Inverse:\n", np.linalg.inv(x))
# Linear algebra operations


print("\n----- Copy vs View -----")
v = a.view()
v[0] = 100
print("View:", v)
print("Original a after view:", a)
# View affects original array

a = np.array([1, 2, 3, 4, 5])  # reset
cp = a.copy()
cp[0] = 200
print("Copy:", cp)
print("Original a after copy:", a)
# Copy does NOT affect original array
```
