```python
'''The syntax of the range function is simple. Different implementations of range function include:

range(n): This creates a range object that has elements from 0 to n-1 [ both inclusive].

range(m,n): This creates a range object with elements from m to n-1 [both inclusive].

range(m,n,t): This creates a range object with elements from m to n-1 with a step count of t.

In other words, the range object has elements m, m+t, m+2t…so on. If t is negative, the elements would be decreasing, and that's exactly what happens in the range (100, 0, -1).'''
```

```python
print(list(range(1,100)))
```

```python
print(list(range(1,100,2)))
```

```python
print(list(range(100,0,-1)))#print  the list in the reverse order
```

```python
list_numbers = (1,2,3,4)
sample_map = map(lambda x: x*2, list_numbers)
print(list(sample_map))
```

```python
mapcube=map(lambda x:x*x*x,list_numbers)
print(mapcube)#Here it retuns a map object we need to type cast into another collection format
print(list(mapcube))
```

```python
#If we are creating a traditional function then this is how it gets passed to the map function for further processing
def multi(x):
    return x*2

list_numbers = [1,2,3,4]
sample_map = map(multi, list_numbers)

print(list(sample_map))
```

```python
#Filter Function
```

```python
'''Filter' is a similar operation, but it requires the function to look for a condition and then returns only those elements from the collection that satisfy the condition.
'''
```

```python
#Note the function passsed to the filter function is always a boolean function
```

```python
#Suppose we have a list where we have mentioned the name of the country
listCon=["India","Israel","Japan","Cyprus","Mauritius"]
```

```python
#Lets apply all the three functions and understand waht is the work of all of these Functions
```

```python
#Use Map function to chnage the letters name of the countries to uppercase i.e we are making changes in the elements of the list(or collection)
mapex=map(lambda x:x.upper(),listCon)
print(list(mapex))
```

```python
#Use filter function to check a condition if its present in the certain elements in the list or collection or not
#if yes then what are the elemts that satisfies that condition
filterex=filter(lambda x:x.startswith("I"),listCon)
print(list(filterex))
```

```python
#Then lets use the reduce function to check
#THE reduce function reduces the collection(list) to concatenate the country name into a single string variable
#this function is present in functools module we need to import that.
#Always takes functions that atleast have 2 parameters/Arguments
```

```python
#Example
from functools import reduce
```

```python
reduceex=reduce(lambda x,y:x+" "+y,listCon)
print(reduceex)
```

```python
reduceex=reduce(lambda x,y:x+" "+y,listCon)
print(list(reduceex))
```

```python
#iN THE ABOVE 2 FUNCTION JUST AFTER CONVERTING THE REDUCEEX TO LIST IT PRODUCES DIFFERENT OUTPUT THAN THE PRINT(REDUCEX) STATEMNT
#The differenece is because the return object type of the reducefunction we used here is string
#Then in the second line of code when we are converting the sting to a list using this list().
#listfunction takes it as a single string then treats its eachletter as an element of the string and returns a list full of letters
```

```python
#notes
'''
#reduce(function, iterable object):

1.The function object passed to the reduce function decides what expression is passed to the iterable object.
2.The iterable object can be a string, list, tuple, set or dictionary.
3.The 'reduce' function produces a single output.

'''
```
